```
from bs4 import BeautifulSoup
import requests
import pandas as pd
from google.colab import files
from csv import writer

html_doc = requests.get ("https://affiliate.rakuten.co.jp/link/high_ranking?l-id=affiliate_PC_top_regular_a1").text
soup = BeautifulSoup(html_doc, 'html.parser') # BeautifulSoupの初期化
```

```
#website からの引用
lists = soup.find_all(class_="item-review")

#列の設定
columns = ["title", "price", "ratio", "link"]
df = pd.DataFrame(columns=columns)

for list in lists:
  # get only text by using get_text()
  title = list.find("a").get_text()  
  price = list.find("li", {"list-inline-item pink-text"}).get_text()
  #to get the next sibling of the indicated item
  ratio = list.find("li", {"list-inline-item pink-text"}).find_next_sibling().get_text()
  link = list.find("a")["href"]
  
  #それぞれのデータを行にはめ込んでいく
  se = pd.Series([title, price, ratio, link], columns)
  print (se)
  df = df.append(se, columns)
  

filename = "rakuten_ranking.csv"
df.to_csv(filename, encoding = 'utf-8-sig') #encoding指定しないと、エラーが起こります。おまじないだともって入力します。
files.download(filename)
```

```
#youtube からの引用
lists = soup.find_all(class_="item-review")

with open ("rakuten.csv","w") as csv_file:
  csv_writer = writer(csv_file)
  headers = ["Title", "Price", "Link"]
  csv_writer.writerow(headers)
  
  for list in lists:
    title = list.find("a").get_text()  
    price = list.find("li", {"list-inline-item pink-text"}).get_text()
    link = list.find("a")["href"]
    csv_writer.writerow([title, price, link])
```
